```python
X = df.drop(columns=['loan_paid_back'])
y=df[["loan_paid_back"]]
```

```python

# ColumnTransformer
preproc = make_column_transformer(
    # Log-transform annual_income
    (make_pipeline(SimpleImputer(), FunctionTransformer(np.log1p, validate=True), RobustScaler()), ['annual_income']),

    # Scale all other numeric columns automatically
    (make_pipeline(SimpleImputer(), RobustScaler()),
     make_column_selector(dtype_include='float64')),

    # One-hot encode categorical columns automatically
    (make_pipeline(OneHotEncoder(handle_unknown='ignore')),
     make_column_selector(dtype_include='object')),

    remainder='passthrough'
)

# Fit-transform
X_processed = preproc.fit_transform(X)
print("Processed shape:", X_processed.shape)
```

# Recommended Modeling Workflow

1. **Start with a baseline model (Logistic Regression)**  
   - Simple, fast, and interpretable  
   - Helps verify that preprocessing works  
   - Provides a baseline performance score

2. **Try a tree-based model (Random Forest)**  
   - Handles nonlinear relationships  
   - More robust on tabular data  
   - Usually performs better than logistic regression

3. **Use a gradient boosting model (LightGBM / XGBoost / CatBoost)**  
   - Often achieves top performance on Kaggle  
   - Very strong for large tabular datasets  
   - Efficient, accurate, and handles feature interactions well

4. **Evaluate all models with proper validation**  
   - Use a train/validation split or cross-validation  
   - Track metrics such as Accuracy, ROC-AUC, Precision/Recall  
   - Avoid overfitting and compare models fairly

5. **Train the final model on the full training dataset and predict on the test set**  
   - Refit preprocessing + model on all training data  
   - Apply preprocessing (`transform`) to test data  
   - Generate predictions and create a Kaggle submission file
